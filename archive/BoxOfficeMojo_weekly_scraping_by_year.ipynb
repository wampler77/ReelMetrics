```python
# packages to import
import pandas as pd
import bs4
import random
from bs4 import BeautifulSoup # BeautifulSoup
import requests
from io import StringIO
from time import sleep
```

```python
weeklyTables = {} # Where all of our movie weekly gross tables will end up
for year in range(1977, 2025):
    # getting the webpage info and making sure we didn't get an error.
    response = requests.get(url="https://www.boxofficemojo.com/year/"+str(year)+"/")
    if int(response.status_code) == 200:
        print("Request successful!"+'\n')
    else:
        print("Request Unsuccessful!"+'\n')

    yearLinks = [] # Where each link to a yearly top movie page will be stored
    # Creating a BeautifulSoup object
    # This takes a Year link and generates a list of links to all movies listed
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find_all('table')[0] 
    releases = table.find_all("td", attrs={"class":"a-text-left mojo-field-type-release mojo-cell-wide"})
    for a in releases:
        yearLinks.append(a.find("a",attrs={"class":"a-link-normal"})['href'])

    # This takes each release from the year and makes a data frame from the weekly gross table
    # 
    for yearLink in yearLinks:
        sleep(random.uniform(0, 0.1))
        movieLink = yearLink.split('?')[0]
        movieUrl = "https://www.boxofficemojo.com"+movieLink+"weekly/"
        response = requests.get(url=movieUrl)
        if int(response.status_code) == 200:
            print("Request successful!"+'\n')
        else:
            print("Request Unsuccessful!"+'\n')
        weekLinks = []
        # Creating a BeautifulSoup object
        movieSoup = BeautifulSoup(response.text, 'html.parser')
        try:
            df = pd.read_html(StringIO(str(movieSoup.find('table'))))[0]
            movieName = movieSoup.find("h1", attrs={"class":"a-size-extra-large"}).text
            weeklyTables[movieName] = df
        except:
            print("Table not found")

```

```python
import pickle
pickle.dump( weeklyTables, open( "BoxOfficeMojoWeekly.p", "wb" ) )
```
